##Cálculo Média Anual Fiap

```python
# Primeiro Semestre
# Cálculo de Média dos Checkpoints
# 3 CPs obrigatórios ----> input
# Eliminar a menor nota
# Calcular a média das duas maiores notas

CP1, CP2, CP3 = input("Insira as notas dos checkpoints: ").split()
CP1 = float(CP1)
CP2 = float(CP2)
CP3 = float(CP3)

# Descobrir qual a menor nota
menorNotaCPs1 = min(CP1, CP2, CP3)

# realizar o cálculo da média das CPs do 1° semestre
mediaCP_semestre1 = (CP1 + CP2 + CP3 - menorNotaCPs1) / 2

print(mediaCP_semestre1)
```

```python
# Cálculo de Notas de SPRINTs (SP1 e SP2)
SP1, SP2 = input("Insira as notas das SPRINTs: ").split()
SP1 = float(SP1)
SP2 = float(SP2)

# Média das duas notas
mediaSP_semestre1 = (SP1 + SP2)/2
print(mediaSP_semestre1)
```

```python
# Cálculo de notas de Global Solution (GS)
# Insira a nota
notaGS1 = (input("Insira nota da Global Solution(ou precione enter caso não houver nota ): "))

# Caso não haja nota de GS, o aluno deverá inserir sua nota de SUB

if not notaGS1:
  print("Nota da Global Solution não inserida.Insira a nota da substitutiva... ")
  notaGS1 = float(input("Insira nota da Substitutiva: "))
else:
  notaGS1 = float(notaGS1)
```

```python
# Cálculo da Média Semestral 1

media_semestral1 = 0.6*notaGS1 + 0.2*mediaCP_semestre1 + 0.2*mediaSP_semestre1
print("Média do Primeiro Semestre: ",media_semestral1)
```

```python
# Segundo Semestre
# Cálculo de Média dos Checkpoints
# 3 CPs obrigatórios ----> input
# Eliminar a menor nota
# Calcular a média das duas maiores notas

CP4, CP5, CP6 = input("Insira as notas dos checkpoints: ").split()
CP4 = float(CP4)
CP5 = float(CP5)
CP6 = float(CP6)

# Descobrir qual a menor nota
menorNotaCPs2 = min(CP4, CP5, CP6)

# realizar o cálculo da média das CPs do 1° semestre
mediaCP_semestre2 = (CP4 + CP5 + CP6 - menorNotaCPs1) / 2

print(mediaCP_semestre2)
```

```python
# Cálculo de Notas de SPRINTs (SP1 e SP2)
SP3, SP4 = input("Insira as notas das SPRINTs: ").split()
SP3 = float(SP3)
SP4 = float(SP4)

# Média das duas notas
mediaSP_semestre2 = (SP3 + SP4)/2
print(mediaSP_semestre2)
```

```python
# Cálculo de notas de Global Solution (GS)
# Insira a nota
notaGS2 = (input("Insira nota da Global Solution(ou precione enter caso não houver nota ): "))

# Caso não haja nota de GS, o aluno deverá inserir sua nota de SUB

if not notaGS2:
  print("Nota da Global Solution não inserida.Insira a nota da substitutiva... ")
  notaGS2 = float(input("Insira nota da Substitutiva: "))
else:
  notaGS2 = float(notaGS1)
```

```python
# Cálculo da Média Semestral 2

media_semestral2 = 0.6*notaGS2 + 0.2*mediaCP_semestre2 + 0.2*mediaSP_semestre2
print("Média do Segundo Semestre: ",media_semestral2)
```

```python
# Média Final Fiap
mediaFinal = 0.4*media_semestral1 + 0.6*media_semestral2
print("Nota Media Final: ", mediaFinal)
```

```python
# Apresentar se o aluno está aprovado, reprovado ou de exame.
# Aprovado >= 6.0
# Exame >= 4.0 e < 6.0
# Reprovado < 4.0

if mediaFinal >= 6.0:
  print("Aprovado")
elif mediaFinal < 6.0 or mediaFinal >=4:
  print("O Aluno Está de Exame.Insira a nota do Exame...")
  notaExame = float(input("Insira nota Exame: "))
  mediaDefinitiva = (notaExame + mediaFinal)/2
  if mediaDefinitiva >= 6.0:
    print("Aluno foi Aprovado")
  else:
    print("Aluno Reprovou")

else:
  print("O Aluno Reprovou")



```
